```python
from sklearn import datasets as ds
import matplotlib.pyplot as plt
from sklearn import cluster as ct
import sklearn.metrics as mt
```

```python
X, y = ds.make_blobs(
    n_samples=100,
    n_features=2,
    centers=4,
    cluster_std=[0.2, 0.2, 0.9, 0.3],
    random_state=42
)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
n_clusters=4

kmeans = ct.KMeans(
    n_clusters=n_clusters,
    init='random',
    n_init=10,
    random_state=42
    ) 


labels = kmeans.fit_predict(X)
sil_avg = mt.silhouette_score(X, labels)

print('For n_cluster = {}, the average silhouette_score = {}'.format(n_clusters, sil_avg))
```

```python
kmeans.labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
```

# Fine Tunning

```python
n_clusters=[i for i in range(2,11)]
n_list = list()

for i in n_clusters:
    kmeans = ct.KMeans(
        n_clusters=i,
        init='random',
        n_init=10,
        random_state=42
        ) 


    labels = kmeans.fit_predict(X)
    sil_avg = mt.silhouette_score(X, labels)
    n_list.append(sil_avg)

    print('For n_cluster = {}, the average silhouette_score = {}'.format(i, sil_avg))
```

```python
plt.plot(n_clusters, n_list, marker='o')
plt.xlabel('Número de Clusters K')
plt.ylabel('Silhouette Score Average')
```
